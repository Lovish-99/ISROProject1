```python
#importing libraries and modules
import time
import datetime
import hashlib
from urllib.request import urlopen, Request
from bs4 import BeautifulSoup
import pandas as pd
from time import gmtime, strftime
import os
import csv
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
def element_name():
    master_list = ["Dissolved Oxygen","Bio Chemical Oxygen Demand","Total Coliform","pH"]
    return master_list
```

```python
#function for creating the directories at the specified locations 
def create_dir():
    master_list = element_name()
    directory = "Runtime2"
    parent_dir = "D:/"
    path = os.path.join(parent_dir, directory)    
    #create a main directory
    os.mkdir(path)

    for i in master_list:
        directory = i
        parent_dir = "D:/Runtime2/"
        path = os.path.join(parent_dir, directory)
        
        #create a main directory
        os.mkdir(path)

        for j in range(1,93):
            directory1 = "Location" + str(j)
            parent_dir = "D:/Runtime2/" + i
            path = os.path.join(parent_dir, directory1)
            
            #create a sub-directory of a main directory
            os.mkdir(path)
#calling of create_dir function done
#create_dir()
```

```python
#function for generating the location and data csv files for the first time use 
def create_csv():
    #getting the url
    url = urlopen("http://125.19.52.219/wqi/")

    #creating BeautifulSoup Object
    soup = BeautifulSoup(url, "html.parser")

    #get the required data using soup object fron script tag
    scrt = soup.find_all("script")
    #converting the script tag data in string and replacing '\n' & '\t' tags
    final_str=str(scrt[1]).replace("\n","")
    final_str=final_str.replace("\t","")
    
    #create empty list
    master_list = []
    
    #create datetime object
    now = datetime.datetime.now()
    
    #extracting required data using string index method
    for i in range(1,93):
        #creating empty dictionaries
        temp_dic1 = {}
        
        #creating empty lists
        master_list2 = []
       
        name = element_name()
        for el in name:
            temp_dic2 = {}
            master_list1 = []
            #finding the indicies of the required data from the scraped data
            index=final_str.index("google.maps.LatLng(")
            index2 = final_str.index('class="head">')
            index3 = final_str.index("Monitoring")
            index4 = final_str.index("Sampling Month")
            index5 = final_str.index(el)


            #--------------------------------------setting values of dictionary----------------------------------------------
            temp_dic2["Frequency Of Monitoring"] = final_str[index3+11:index3+30].split("<")[0]
            temp_dic2["Sampling Month"] = final_str[index4+15:index4+23]
            temp_dic2["Date"] = now.strftime('%d %b %Y')
            temp_dic2["Time"] = now.strftime('%H:%M:%S')
            s = len(el) + 2
            temp_dic2[el] = final_str[index5+s:index5+40].split("<")[0]

            #appending the dictionary in master list1
            master_list1.append(temp_dic2) 
            #create pandas DataFrame from master-list
            df = pd.DataFrame(master_list1)
            #creating a csv file of the data
            df.to_csv('D:/Runtime2/{}/Location{}/Data.csv'.format(el,str(i)),index=True)
        
        
        #-------------------------------------setting values for the location file----------------------------------------
        temp_dic1["Location's"] = "Location-" + str(i)
        temp_dic1["Name Of The River"] = final_str[index2+13:index2+90].split("<")[0]
        temp_dic1["Latitude"] = final_str[index+19:index+27]
        temp_dic1["Longitude"] = final_str[index+30:index+39]

        #changing the value of index in order to extract the next location data
        final_str=final_str[index+300:]
        master_list.append(temp_dic1)
        
    #create pandas DataFrame from master-list
    lc = pd.DataFrame(master_list)
    #generate csv file of the location names
    lc.to_csv("D:/Runtime2/Location's.csv",index=True)

#calling the create_csv function
#create_csv()
```

```python
def create_graph():
    masterlist6 = element_name()
    for i in masterlist6:
        for loc in range(1,93):
            df = pd.read_csv("D:/Runtime2/{}/Location{}/Data.csv".format(i,str(loc)))
            g = sns.barplot(x = "Time", y = i , data = df)
            labels = g.get_xticklabels()
            plt.title('Analysis of Turbidity of Water')
            plt.xlabel("Time")
            plt.setp(labels, rotation = 20, horizontalalignment = 'right')
            plt.savefig('D:/Runtime2/{}/Location{}/Data.jpeg'.format(i,str(loc)),dpi=300)
#create_graph()
```

```python
# setting the URL we want to monitor
url = Request('http://125.19.52.219/wqi/', headers={'User-Agent': 'Mozilla/5.0'})

print("running")

#monitering the website
while True:
    try:
        #open file in read mode and get the old hashcode of the site
        file = open("hashcode.txt","r")
        currentHash = file.readline()
        file.close()
        
        # perform the get request
        response = urlopen(url).read()
          
        # create a new hash
        newHash = hashlib.sha224(response).hexdigest()
        
        # check if new hash is same as the previous hash
        if newHash == currentHash:
            print("Data not updated")
            
            # wait for 10 seconds
            time.sleep(3600)
            continue
  
        # if something changed in the hashes
        else:
            # notify
            print("Data updated")
            create_csv2()
            create_graph()
            
            #create a file, write it and close it
            file = open("hashcode.txt","w+")
            file.write(newHash)
            file.close()

            # wait for 10 seconds
            time.sleep(3600)
            continue
        
    #handling exceptions
    except Exception as e:
        print("error",e)
        continue
```
