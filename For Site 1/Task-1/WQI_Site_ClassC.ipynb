```python
#importing required packages and modules 
from bs4 import BeautifulSoup
from urllib.request import urlopen, Request
import pandas as pd
from time import gmtime, strftime
import hashlib

#getting the url
html = urlopen("http://125.19.52.219/wqi/bathing.php")

def create_hash():
    # setting the URL we want to monitor
    url = Request('http://125.19.52.219/wqi/bathing.php', headers={'User-Agent': 'Mozilla/5.0'})

    #get the url and read it
    response = urlopen(url).read()
    currentHash = hashlib.sha224(response).hexdigest()
    
    #create a file, write it and close it
    file = open("hashcode_3.txt","w+")
    file.write(currentHash)
    file.close()

#creating BeautifulSoup Object
soup = BeautifulSoup(html, "html.parser")

#get the required data using soup object fron script tag
scrt = soup.find_all("script")

#converting the script tag data in string and replacing '\n' & '\t' tags
final_str=str(scrt[1]).replace("\n","")
final_str=final_str.replace("\t","")

#create empty list
master_list = []

#extracting required data using string index method
for i in range(92):
    #create empty dictionary
    temp_dic = {}
    
    #searching the index from the string-script
    index=final_str.index("google.maps.LatLng(")
    index2 = final_str.index('class="head">')
    index3 = final_str.index("Monitoring")
    index4 = final_str.index("Sampling Month")
    index5 = final_str.index("Dissolved Oxygen")
    index6 = final_str.index("Bio Chemical Oxygen Demand")
    index7 = final_str.index("Fecal Coliform")
    index8 = final_str.index("pH")
    
    #adding details to dictionary
    temp_dic["Location"] = final_str[index2+13:index2+90].split("<")[0]
    temp_dic["Latitude"] = final_str[index+19:index+27]
    temp_dic["Longitude"] = final_str[index+30:index+39]
    temp_dic["Frequency Of Monitoring"] = final_str[index3+11:index3+22]
    temp_dic["Sampling Month"] = final_str[index4+15:index4+23]
    temp_dic["Dissolved Oxygen"] = final_str[index5+18:index5+30].split("<")[0]
    temp_dic["Bio Chemical Oxygen Demand"] = final_str[index6+28:index6+40].split("<")[0]
    temp_dic["Fecal Coliform"] = final_str[index7+16:index7+35].split("<")[0]
    temp_dic["pH"] = final_str[index8+4:index8+14].split("<")[0]
    
    #changing the index for the next search
    final_str=final_str[index+300:]
    
    #adding temporary dictionary in main list
    master_list.append(temp_dic)
    
#create pandas DataFrame from master-list
df = pd.DataFrame(master_list)

#generating csv file of the scraped data
df.to_csv('WaterQualityDataC_{}.csv'.format(strftime("%d %b %Y %H %M %S", gmtime())),index=True)

#displaying the scraped data
display(df)

#calling create_hash function to create a hash code for this generated data 
create_hash()
```
